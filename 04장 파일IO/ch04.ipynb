> Q1
``` py
    주어진 자연수가 홀수인지 짝수인지 판별해 주는 함수(is_odd)를 작성해 보자.
```

````python
def is_odd(number):
    if number % 2 ==1:
        return True
    else:
        return False

n = 8
# is_odd(n)
if is_odd(n):
    print(f"{n}은(는) 홀수입니다.")
else:
    print(f"{n}은(는)짝수입니다.")


# print(is_odd(n))
````

> Q2
``` py
    입력으로 들어오는 모든 수의 평균 값을 계산해 주는 함수를 작성해 보자. 
    (단 입력으로 들어오는 수의 개수는 정해져 있지 않다.)
    ※ 평균 값을 구할 때 len 함수를 사용해 보자.
```

````python
def get_avg(args):
    return sum(args) / len(args)

data_in = 10.3, 12.1, 30, 3.14, 10
type(data_in) # 타입은 tuple 임. > 괄호 생략이 가능!


print("평균 = ", round(get_avg(data_in), 2))
# round :: 출력되는 소숫점 자리 지정 
````

> Q5
``` py
    다음은 "test.txt"라는 파일에 "Life is too short" 문자열을 저장한 후 
    다시 그 파일을 읽어서 출력하는 프로그램이다.

    f1 = open("test.txt", 'w')
    f1.write("Life is too short")

    f2 = open("test.txt", 'r')
    print(f2.read())
    이 프로그램은 우리가 예상한 "Life is too short"라는 문장을 출력하지 않는다. 
    우리가 예상한 값을 출력할 수 있도록 프로그램을 수정해 보자.
```

````python
try:
    f1 = open("test.txt", 'w') # write
# open을 하면 무조건 close를 해야함
# 까먹고 안닫는 경우가 많으니. open을 쓰면 바로 close하는 버릇을 들이자.
    f1.write("Life is too short")
except Exception as ex:
    print(ex)   
finally:
    f1.close()

f2 = open("test.txt", 'r') # read
print(f2.read())
f2.close()
````

> Q6
``` py
    사용자의 입력을 파일(test.txt)에 저장하는 프로그램을 작성해 보자. 
    (단 프로그램을 다시 실행하더라도 기존에 작성한 내용을 유지하고 새로 입력한 내용을 추가해야 한다.)
```

````python
# 사용자에게 메시지를 입력받아 기존 test에 저장해보자
dt = input("메시지를 입력해주세요.: ")
# 기존 파일에 추가 메세지 저장
# encoding = 'utf-8' : 글씨가 깨져보일 수 있으니 추가한 부분.

with open("test.txt", 'a', encoding = 'utf-8') as fw: # close 안해도 됨
    fw.write('\n' + dt)

with open("test.txt", 'r', encoding = 'utf_8') as fr:
    print(fr.read)
````

> Q7
``` py
    다음과 같은 내용을 지닌 파일 test.txt가 있다. 
    이 파일의 내용 중 "java"라는 문자열을 "python"으로 바꾸어서 저장해 보자.

    Life is too short
    you need java
    ※ replace 함수를 사용해 보자.
```

````python
with open("test.txt", 'rt', encoding = 'utf-8') as fr:
    data = fr.read().replace("java", "python")

with open("test.txt", 'wt', encoding = 'utf-8') as fw:
    fw.write(data)
````
